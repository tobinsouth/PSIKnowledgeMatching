```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
rounding_df = pd.read_csv("rounding_recalls_quarter.csv")
noise_df = pd.read_csv("noise_recall_quarter.csv")


# Create the plot for Recall@100
plt.figure()
plt.plot(rounding_df["Rounding_Decimals"], rounding_df["Recall@100"], label="Rounding")
plt.plot(noise_df["Rounding_Decimals"], noise_df["Recall@100"], label="Noise")


# Determine the value for the 12th Rounding_Decimals
baseline = rounding_df.loc[rounding_df["Rounding_Decimals"] == 12, "Recall@100"].iloc[0]
# Add a vertical dotted line for the baseline at the 12th Rounding_Decimals value
plt.axhline(y=baseline, color='gray', linestyle='--', linewidth=1, label="Baseline")

# Customize plot labels and title
plt.xlabel("Rounding Decimals")
plt.ylabel("Recall@100")
plt.title("Recall@100 vs Rounding Decimals with Quarter of Data")
plt.legend()

# Show the plot
plt.show()
```
