```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
rounding_df = pd.read_csv("rounding_recalls_quarter.csv")
noise_df = pd.read_csv("noise_recall_quarter.csv")
```

```python
# Create the first plot for Recall@100
plt.figure()
plt.plot(rounding_df["Rounding_Decimals"], rounding_df["Recall@100"], label="Rounding")
plt.plot(noise_df["Rounding_Decimals"], noise_df["Recall@100"], label="Noise")
plt.xlabel("Rounding Decimals")
plt.ylabel("Recall@100")
plt.title("Recall@100 vs Rounding Decimals with Quarter of data")
plt.legend()
plt.show()

```

```python
# Create the second plot for Recall@1000
plt.figure()
plt.plot(rounding_df["Rounding_Decimals"], rounding_df["Recall@1000"], label="Rounding")
plt.plot(noise_df["Rounding_Decimals"], noise_df["Recall@1000"], label="Noise")
plt.xlabel("Rounding Decimals")
plt.ylabel("Recall@1000")
plt.title("Recall@1000 vs Rounding Decimals with Quarter of data")
plt.legend()
plt.show()
```
